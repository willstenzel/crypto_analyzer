```python
import dataExtractor
import pandas as pd
```

```python
Data_df = (dataExtractor.createCompleteDataset("BTC-USD", "2007-12-31", "2017-12-31", 10))
coin_data = (dataExtractor.getCoinData("BTC-USD", "2007-12-31", "2017-12-31"))
close_price_df = pd.DataFrame(coin_data['Adj. Close'])
```

```python
Data_df
```

```python
pct_change_df = dataExtractor.pctChangeData("BTC-USD", "2007-12-31", "2017-12-31")
pct_change_df = dataExtractor.removeDifferentDates(pct_change_df, Data_df)[0]
Data_df = dataExtractor.removeDifferentDates(pct_change_df, Data_df)[1]
pct_change_df
```

```python
plottingFrame = pd.concat([Data_df, pct_change_df], axis = 1)
plottingFrame
```

```python
close_price_df = pd.DataFrame(coin_data['Adj. Close'])
close_price_df = dataExtractor.removeDifferentDates(close_price_df, plottingFrame)[0]
plottingFrame = dataExtractor.removeDifferentDates(pct_change_df, plottingFrame)[1]
close_price_df
```

```python
plottingFrame = pd.concat([plottingFrame, close_price_df], axis = 1)
plottingFrame
```

Now we've done all the manipulations necessary to get a dataframe we can plot. For the dataframe that will go into
the machine learning model, the 'Pct. Change' and 'Adj. Close' columns won't be present (the Dataframe will end at 
'Up or Down?', as that is the label.)

Below, you an see Close Prices, Percent Change, RSI, and Momentum plotted over time. Interesting observations from these graphs are:
1) There's a massive spike in Percent Change in late 2013 that's more attributable to the low price of Bitcoin than it is to actual financial fundamentals, this might be an area that could be a set of outliers to remove. 
2) RSI and Percent Change both seem to be cyclical and associated, while Price and Momentum associate strongly. Then again, Momentum seems to track retroactively, while RSI looks proactively cyclical. This may mean that RSI is a stronger indicator of positive/negative returns. 

```python
ax1 = plottingFrame.plot(kind = 'line', y = 'Adj. Close', figsize = (15, 5), title = 'Close Price Over Time')
ax1.set_xlabel("Date")
ax1.set_ylabel("Price")
ax2 = plottingFrame.plot(kind = 'line', y = 'Pct. Change', figsize = (15, 5), title = 'Pct. Change Over Time')
ax2.set_xlabel("Date")
ax2.set_ylabel("Percent Change")
ax3 = plottingFrame.plot(kind = 'line', y = 'RSI', figsize = (15, 5), title = 'Relative Strength Index Over Time')
ax3.set_xlabel("Date")
ax3.set_ylabel("RSI")
ax4 = plottingFrame.plot(kind = 'line', y = 'MOM', figsize = (15, 5), title = 'Momentum Over Time')
ax4.set_xlabel("Date")
ax4.set_ylabel("Momentum")
```

Below is a plot of the different kinds of moving averages that will be used as features in the algorithm (Kauffman Adaptive, Exponential, Triple Exponential, and Triangular). The graph seems to show that there is greater price sensitivity in the Triple Exponential MA and that peaks seem to show earlier in the TEMA and EMA graphs. Given this, it may follow that periods of negative returns follow peaks in exponential moving averages. It may be useful to see if we can engineer a feature that tracks the rate of change of these moving averages. 

```python
ax5 = plottingFrame.plot(kind = 'line', y = "MA", figsize = (15, 10), title = "Plotting Different Kinds of Moving Averages Over Time")
ax5.set_xlabel("Date")
ax5.set_ylabel("Average Price Given By MA Type")
plottingFrame.plot(kind = 'line', y = 'KAMA', ax = ax5)
plottingFrame.plot(kind = 'line', y = 'EMA', ax = ax5)
plottingFrame.plot(kind = 'line', y = 'TEMA', ax = ax5)
plottingFrame.plot(kind = 'line', y = 'TRIMA', ax = ax5)
```

Below are two common indicators of volatility, Rate of Change and Beta values. What seems interesting here is that the ROC seems to follow similar trends as the Beta, but peaks and values seem to extend farther and take place more gradually than in the graph of the Beta. This may implicate the Beta value as a trailing price indicator, especially when combined with a cyclical RSI graph. 

```python
ax6 = plottingFrame.plot(kind = 'line', y = 'ROC', figsize = (15, 5), title = "Rate of Change Over Time")
ax6.set_xlabel("Date")
ax6.set_ylabel("ROC")
ax7 = plottingFrame.plot(kind = 'line', y = 'BETA', figsize = (15, 5), title = "Beta Value Over Time")
ax7.set_xlabel("Date")
ax7.set_ylabel("Beta Value")
```

The plot below acts as a proof of sorts that technical indicators for stocks can also be applied to cryptocurrencies. Bollinger bands are intended to create upper and lower bounds that indicate equity volatility, and the graph below shows that the same fundamental formulas can be replicated and applied to a cryptocurrency price chart. 

```python
ax8 = plottingFrame.plot(kind = 'line', y = "Adj. Close", figsize = (15, 10), title = "Plotting Bollinger Bands With Close Price Over Time")
ax8.set_xlabel("Date")
ax8.set_ylabel("Price Level")
plottingFrame.plot(kind = 'line', y = 'Upper Band', ax = ax8)
plottingFrame.plot(kind = 'line', y = 'Middle Band', ax = ax8)
plottingFrame.plot(kind = 'line', y = 'Lower Band', ax = ax8)
```

The graph below shows the parabolic SAR value of Bitcoin over time. Parabolic SAR (the abbreviation for Parabolic Stop-And Reverse). Parabolic SAR is intended to indicate reversal points for equity prices, thereby allowing investors to predict selloffs or strong buying periods. The graph below does not seem to dispute this notion, as it also seems to trail the price graph just slightly. 

```python
ax9 = plottingFrame.plot(kind = 'line', y = 'SAR', title = "Parabolic SAR Over Time")
ax9.set_xlabel("Date")
ax9.set_ylabel("SAR")
ax10 = plottingFrame.plot(kind = 'line', y = 'Adj. Close', title = "Close Price Over Time")
ax10.set_xlabel("Date")
ax10.set_ylabel("Price")
```
